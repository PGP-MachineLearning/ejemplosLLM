```python
#@title Cargar Librerías

from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch  # nota: usa Torch porque con TF directo tira error
torch.random.manual_seed(0)

import re
import random

import textwrap

print("Librerías cargadas.")

# determina si usa GPU o CPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

print("Device: ", device)
```

# Modelo para Generar Texto ( Text Generation )

```python
#@title Cargar Modelo Text Generation

nombreModeloTG = "microsoft/Phi-4-mini-instruct" #@param [ "microsoft/Phi-4-mini-instruct", "microsoft/phi-2" ]
#@markdown otros modelos disponibles en: https://huggingface.co/models?pipeline_tag=text2text-generation&sort=trending


# cargar modelo
model = AutoModelForCausalLM.from_pretrained(
    nombreModeloTG,
    device_map="auto",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained(nombreModeloTG)



pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}


# ejecuta el modelo para generar el texto
def genText(texto_inicio):
  outputs = pipe(texto_inicio, **generation_args)
  text = outputs[0]['generated_text']
  return text

# geneera el texto y lo muestra
def procesar(text, wrapResult=False):
  res = genText(text)
  print("")
  print(">", textwrap.fill(text, 100))
  print("--> Respuesta: ")
  if wrapResult:
    print(textwrap.fill(res, 100))
  else:
    print(res)
  print("")
```

# Preguntas generales:

```python
#@title Probar responder pregunta 1

texto_inicio = "how should a train a neural network?" #@param{type:"string"}

procesar(texto_inicio, True)
```

```python
#@title Probar responder pregunta 2

texto_inicio = "what is a prime number?" #@param{type:"string"}

procesar(texto_inicio, True)
```

# Generar Código:

```python
#@title Probar generar código 1

code_requirement = "Select all clients that have bought at least a product" #@param{type:"string"}

procesar(code_requirement)
```
