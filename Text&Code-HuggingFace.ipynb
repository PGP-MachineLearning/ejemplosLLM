# Modelo para Generar Texto ( Text Generation )

```python
#@title Cargar Modelo Text Generation

nombreModeloTG = "microsoft/phi-2" #@param [ "microsoft/phi-2" ]
#@markdown otros modelos disponibles en: https://huggingface.co/models?pipeline_tag=text2text-generation&sort=trending


from transformers import AutoModelForCausalLM, AutoTokenizer


# cargar modelo
model = AutoModelForCausalLM.from_pretrained(nombreModeloTG)
tokenizer = AutoTokenizer.from_pretrained(nombreModeloTG)

# ejecuta el modelo para generar el texto
# ojo que si se pone largo muy grande, se queda sin memoria Colab
def genText(texto_inicio, texto_largo_max = 150):
  inputs = tokenizer(texto_inicio, return_tensors="pt", return_attention_mask=False)
  outputs = model.generate(**inputs, max_length=texto_largo_max, pad_token_id=tokenizer.eos_token_id)
  text = tokenizer.batch_decode(outputs)[0]
  return text

# geneera el texto y lo muestra
def procesar(text, wrapResult=False):
  res = genText(text)
  print("")
  print(">", textwrap.fill(text, 100))
  print("--> Respuesta: ")
  if wrapResult:
    print(textwrap.fill(res, 100))
  else:
    print(res)
  print("")
```

```python
#@title Probar responder preguntas

texto_inicio = "cómo puedo entrenar una red neuronal?" #@param{type:"string"}

procesar(texto_inicio, True)
```

```python
#@title Probar generar código 1

code_requirement = "def print_prime(n): \"\"\" Print all primes between 1 and n \"\"\"'''," #@param{type:"string"}

procesar(code_requirement)
```

```python
#@title Probar generar código 2

code_requirement = "def find(x, l): \"\"\" Print all position of value x in list l \"\"\"'''," #@param{type:"string"}

procesar(code_requirement)
```
